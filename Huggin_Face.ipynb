Intalar la librería transformers

```python
# Importa la función pipeline
from transformers import pipeline

# Carga un pipeline para la tarea de 'text-generation'
# Especificamos el modelo gpt2, uno de los primeros modelos de lenguaje a gran escala
generador = pipeline("text-generation", model="gpt2")
```

Importar y Crear el Pipeline de Generación Indicamos a pipeline que la tarea es generación de texto.

```python
# Define el prompt inicial (la primera parte del texto)
prompt = "The future of Artificial Intelligence in medicine will be"

# Genera texto basándose en el prompt
# max_length: Define el número máximo de tokens (palabras/partes de palabras) en la salida.
resultado = generador(prompt, max_length=50, num_return_sequences=1)

# Imprime el texto generado
print(resultado[0]['generated_text'])
```

```python
# Define el prompt inicial (la primera parte del texto)
prompt = "The future of Artificial Intelligence in medicine will be"

# Genera texto basándose en el prompt
# max_length: Define el número máximo de tokens (palabras/partes de palabras) en la salida.
resultado = generador(prompt, max_length=50, num_return_sequences=1)

# Imprime el texto generado
print(resultado[0]['generated_text'])
```

Usamos el generador para que complete una frase inicial (el prompt).[texto del vínculo](https://) El modelo predice la siguiente palabra con base en las anteriores, repitiendo el proceso hasta alcanzar la longitud máxima. max_length=50: Le dice al modelo que la salida completa (incluyendo el prompt) no debe superar los 50 tokens.num_return_sequences=1: Solicita una sola secuencia de texto generada. modelo predice la siguiente palabra con base en las anteriores, repitiendo el proceso hasta alcanzar la longitud máxima.

max_length=50: Le dice al modelo que la salida completa (incluyendo el prompt) no debe superar los 50 tokens.

num_return_sequences=1: Solicita una sola secuencia de texto generada.

resultado[0]['generated_text']: El resultado es una lista de diccionarios. Accedemos al primer elemento y extraemos la clave que contiene el texto completo generado.

```python
# Define el prompt
prompt_creativo = "Coding in Python is enjoyable because"

# Genera múltiples respuestas
# num_return_sequences=3: Pide 3 posibles continuaciones.
# do_sample=True: Habilita el muestreo para introducir aleatoriedad (creatividad).
# temperature=0.9: Controla la "creatividad". Valores más altos (cerca de 1.0) son más aleatorios.
resultados_creativos = generador(
    prompt_creativo,
    max_length=70,
    num_return_sequences=3,
    do_sample=True,
    temperature=0.9
)

# Imprime cada respuesta generada
print("--- Posibles Respuestas Generadas ---")
for i, res in enumerate(resultados_creativos):
    print(f"Opción {i+1}: {res['generated_text']}\n")
```

do_sample=True: En lugar de elegir siempre la palabra con la probabilidad más alta (generación determinista), el modelo elige al azar entre las palabras más probables (generación estocástica).

temperature: Actúa como un regulador de la distribución de probabilidad de las palabras. Una temperatura más alta suaviza la distribución, haciendo que sea más probable elegir palabras menos probables, lo que resulta en texto más creativo o inesperado. Una temperatura baja hace que el modelo sea más seguro y repetitivo.
