```python
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cosine

# Modelo multilingue
model = SentenceTransformer("sentence-transformers/LaBSE")

# Descripcion del puesto
descripcion_puesto = """
Buscamos un Analista de Datos con experiencia en Python, SQL y visualización de datos.
Debe poder comunicar hallazgos de manera efectiva a equipos no técnicos.
"""

# Texto del CV del candidato
cv_candidato = """
Analista de datos con 3 años de experiencia en Python, Power BI y modelado de bases de datos SQL.
Amplias habilidades de comunicación y presentación de resultados.
"""
#cv_candidato = """
#Diseñador gráfico con experiencia en Illustrator, Photoshop y branding.
#Creativo y orientado a la comunicación visual, sin experiencia en análisis de datos.
#"""

# Convertimos a embeddings
emb_puesto = model.encode(descripcion_puesto)
emb_cv = model.encode(cv_candidato)

# Calculamos similitud coseno
similaridad = 1 - cosine(emb_puesto, emb_cv)
print(f"Similitud entre el CV y la vacante: {similaridad:.3f}")
```

Las redes neuronales tienen memoria, es trasformada por medio de trasformers.  

```python
from transformers import pipeline
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cosine

# Traductor español a inglés
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-es-en")

# Modelo de embeddings
matcher = SentenceTransformer("all-MiniLM-L6-v2")

# Texto en español
descripcion_puesto = """
Buscamos desarrollador backend con experiencia en APIs REST, Python y
bases de datos PostgreSQL.
"""
cv_candidato = """
Ingeniero de software especializado en desarrollo backend,
con proyectos en Django y PostgreSQL.
"""
#cv_candidato = """
#Diseñador gráfico con experiencia en Illustrator, Photoshop y branding.
#Creativo y orientado a la comunicación visual, sin experiencia en análisis de datos.
#"""

# Traducimos ambos
desc_en = translator(descripcion_puesto)[0]["translation_text"]
cv_en = translator(cv_candidato)[0]["translation_text"]

print("Traducción descripción:", desc_en)
print("Traducción CV:", cv_en)

# Calculamos embeddings y similitud
emb_puesto = matcher.encode(desc_en)
emb_cv = matcher.encode(cv_en)

similaridad = 1 - cosine(emb_puesto, emb_cv)
print(f"Similitud: {similaridad:.3f}")
```
